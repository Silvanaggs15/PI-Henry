ANALISIS DE SENTIMIENTO

```python
# Función para realizar el análisis de sentimiento y asignar valores numéricos
def sentiment_to_numeric(text):
    if pd.isna(text):
        return 1  # Valor '1' para reseñas ausentes
    analysis = TextBlob(text)
    if analysis.sentiment.polarity < 0:
        return 0  # Reseña mala
    elif analysis.sentiment.polarity == 0:
        return 1  # Reseña neutral
    else:
        return 2  # Reseña positiva

# Aplicar la función a la columna 'user_reviews.review' y crear la nueva columna 'sentiment_analysis'
df_australian_user_reviews['sentiment_analysis'] = df_australian_user_reviews['reviews'].apply(sentiment_to_numeric)

# Eliminar la columna original 'user_reviews.review' si ya no se necesita
df_australian_user_reviews.drop('reviews', axis=1, inplace=True)

# Guardar el DataFrame actualizado en un nuevo archivo CSV 
df_australian_user_reviews.to_csv("australian_user_reviews_with_sentiment.csv", index=False)
```

```python
import pandas as pd

# Rutas a los archivos CSV originales
ruta_steam_games = "steam_games.csv"
ruta_australian_user_items = "australian_user_items.csv"

# Columnas que deseas incluir en el archivo "Consulta_final"
columnas_steam_games = ["id", "genres", "release_date"]
columnas_australian_user_items = ["item_id", "playtime_forever"]

# Leer los datos relevantes de ambos archivos
steam_games = pd.read_csv(ruta_steam_games, usecols=columnas_steam_games)
australian_user_items = pd.read_csv(ruta_australian_user_items, usecols=columnas_australian_user_items)

# Combinar los datos en un nuevo DataFrame
consulta_final = pd.concat([steam_games, australian_user_items], axis=1)

# Guardar el DataFrame en un nuevo archivo CSV
consulta_final.to_csv("Consulta_final.csv", index=False)
```

```python
# Eliminar filas con valores nulos en alguna columna
consulta_final_sin_nulos = consulta_final.dropna(how='any')

# Limitar el DataFrame a 20,000 filas
consulta_final = consulta_final_sin_nulos.head(20000)

# Guardar el DataFrame limitado en un nuevo archivo CSV
consulta_final.to_csv("Consulta_final.csv", index=False)
```

```python
def PlayTimeGenre(genero):
    # Cargar los datos del archivo "Consulta_final.csv"
    consulta_final = pd.read_csv("Consulta_final.csv")

    # Filtrar los juegos relevantes por genero
    juegos_relevantes = consulta_final[consulta_final["genres"].str.contains(genero, case=False, na=False)]

    # Agrupar por anio y sumar las horas jugadas
    grouped_data = juegos_relevantes.groupby(juegos_relevantes["release_date"].str[:4])["playtime_forever"].sum().reset_index()

    # Encontrar el anio con más horas jugadas
    max_playtime_year = grouped_data.loc[grouped_data["playtime_forever"].idxmax()]["release_date"]

    return {"Anio de lanzamiento con más horas jugadas para el genero": int(max_playtime_year)}
```

 Creación de la Función de Recomendación: crearemos una función de recomendación basada en la similitud de coseno entre "user_id" y "item_id". Usaremos el archivo "user_juegos.csv" que hemos creado en la Parte 1.

```python
from sklearn.metrics.pairwise import cosine_similarity

def recomendacion_usuario(usuario_id):
    # Cargar el archivo "user_juegos.csv"
    df = pd.read_csv("user_juegos.csv")

    # Filtrar el DataFrame para obtener el usuario dado
    usuario = df[df['user_id'] == usuario_id]

    if usuario.empty:
        return {"message": "Usuario no encontrado"}

    # Eliminar la columna de usuario temporalmente para calcular la similitud del coseno
    df_sin_usuario = df[df['user_id'] != usuario_id]

    # Calcular la similitud del coseno entre el usuario dado y todos los demás usuarios
    similitud = cosine_similarity(usuario[['user_id', 'item_id']].values, df_sin_usuario[['user_id', 'item_id']].values)

    # Obtener los índices de los usuarios más similares
    usuarios_similares_indices = similitud[0].argsort()[-5:][::-1]

    # Obtener los juegos preferidos por los usuarios similares
    juegos_recomendados = []
    for indice in usuarios_similares_indices:
        juegos_usuario_similar = df_sin_usuario[df_sin_usuario['user_id'] == df.iloc[indice]['user_id']]['item_name'].values
        juegos_recomendados.extend(juegos_usuario_similar)

    # Filtrar juegos que el usuario ya haya jugado
    juegos_recomendados = [juego for juego in juegos_recomendados if juego not in usuario['item_name'].values]

    # Tomar los primeros 5 juegos recomendados
    juegos_recomendados = juegos_recomendados[:5]

    return {"Recomendaciones para el usuario": usuario_id, "juegos_recomendados": juegos_recomendados}
```
